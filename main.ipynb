```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generate a cube full of random numbers with mean=0, variance=1, 50*50*50 values
data = np.random.randn(50, 50, 50)

print("mean: " + str(np.mean(data)))
print("std: " + str(np.std(data)))
```

```python
# Plot 2 sets of data, each with different colour and label
# "r" stands for red, "b" for blue, ... 
# "o" means markers are dots, "x" crosses, ...

plt.plot(data[:, 0, 0], "ro", label="First set")
plt.plot(data[:, 1, 0], "bo", label="Second set")

# Adds a legend box to the figure with the info from "label"
plt.legend()    

# Title and axis names
plt.title("This is a plot")
plt.xlabel("This is the abscissa")
plt.ylabel("This is the ordinate axis")

# Open the figure
plt.show()
```

```python
# Represent 2D data as an image
plt.imshow(data[..., 0], interpolation='none')

# Add the color bar on the right 
plt.colorbar()

# Open the figure
plt.show()
```

```python
# Multiple subplots in the same figure
# The argument for subplot function is: number of rows, number of columns, index of subplot

plt.subplot(131)
plt.imshow(data[..., 0])

plt.subplot(132)
plt.imshow(data[..., 1])

plt.subplot(133)
plt.imshow(data[..., 2]) 

# Open the figure
plt.show()
```

```python
# Plot a set of data against another one
plt.scatter(data[:, 0, 0], data[:, 1, 0])

# Open the figure
plt.show()
```

```python
# Draw a histogram of the data with 10 bins
plt.hist(data[:, 0, 0], bins=10)

# Open the figure
plt.show()
```

```python
# Pot 3D data

from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()

ax = Axes3D(fig)
ax.scatter(data[:, 0, 0], data[:, 1, 0], data[:, 2, 0])

# Open the figure
plt.show()
```

```python
# Prettify the first plot with seaborn (you only need to import it and run "set()")
import seaborn as sns
sns.set()

plt.plot(data[:, 0, 0], "ro", label="First set")
plt.plot(data[:, 1, 0], "bo", label="Second set")
plt.legend()    
plt.title("This is a plot")
plt.xlabel("This is the abscissa")
plt.ylabel("This is the ordinate axis")

# Open the figure
plt.show()

# Go back to using Matplotlib's defaults
sns.reset_orig()
```

```python
# Do some math with the data

set1 = data[:, 0, 0]

set1_diff = np.diff(set1)

plt.subplot(211)
plt.plot(set1, "x")
plt.subplot(212)
plt.plot(set1_diff, "x")

plt.show()
```

```python
# Bonus: draw xkcd-like plots (taken from here: https://matplotlib.org/xkcd/examples/showcase/xkcd.html)

with plt.xkcd():

    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    plt.xticks([])
    plt.yticks([])
    ax.set_ylim([-30, 10])

    data = np.ones(100)
    data[70:] -= np.arange(30)

    plt.annotate(
        'THE DAY I REALIZED\nI COULD COOK BACON\nWHENEVER I WANTED',
        xy=(70, 1), arrowprops=dict(arrowstyle='->'), xytext=(15, -10))

    plt.plot(data)

    plt.xlabel('time')
    plt.ylabel('my overall health')

    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.bar([-0.125, 1.0-0.125], [0, 100], 0.25)
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.set_xticks([0, 1])
    ax.set_xlim([-0.5, 1.5])
    ax.set_ylim([0, 110])
    ax.set_xticklabels(['CONFIRMED BY\nEXPERIMENT', 'REFUTED BY\nEXPERIMENT'])
    plt.yticks([])

    plt.title("CLAIMS OF SUPERNATURAL POWERS")

    plt.show()
    
plt.rcdefaults()
```
