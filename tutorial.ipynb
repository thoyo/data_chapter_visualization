```python
import matplotlib.pyplot as plt
import numpy as np

```

```python
data = np.random.randn(50, 50, 50)
```

```python
plt.plot(data[:, 0, 0], 'ro')
plt.show()
```

```python
plt.imshow(data[..., 0], interpolation='none')
plt.colorbar()

plt.show()
```

```python
plt.subplot(131)
plt.imshow(data[..., 0])
plt.subplot(132)
plt.imshow(data[..., 1])
plt.subplot(133)
plt.imshow(data[..., 2]) 
plt.colorbar()

plt.show()
```

```python
plt.scatter(data[:, 0, 0], data[:, 1, 0])
plt.show()
```
